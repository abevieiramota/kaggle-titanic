**VARIABLE DESCRIPTIONS:**  
- **survival**        Survival
                (0 = No; 1 = Yes)
- **pclass**          Passenger Class
                (1 = 1st; 2 = 2nd; 3 = 3rd)
- **name**            Name
- **sex**             Sex
- **age**             Age
- **sibsp**           Number of Siblings/Spouses Aboard
- **parch**           Number of Parents/Children Aboard
- **ticket**          Ticket Number
- **fare**            Passenger Fare
- **cabin**           Cabin
- **embarked**        Port of Embarkation
                (C = Cherbourg; Q = Queenstown; S = Southampton)

**SPECIAL NOTES:**  
***
**Pclass** *is a proxy for socio-economic status (SES)*  
 1st ~ Upper; 2nd ~ Middle; 3rd ~ Lower  
***
**Age** *is in Years; Fractional if Age less than One (1)*    
 If the Age is Estimated, it is in the form xx.5  
***  
With respect to the family relation variables (i.e. sibsp and parch)  
some relations were ignored.  The following are the definitions used  
for sibsp and parch.  
***
Sibling:  Brother, Sister, Stepbrother, or Stepsister of Passenger Aboard Titanic  
Spouse:   Husband or Wife of Passenger Aboard Titanic (Mistresses and Fiances Ignored)  
***
Parent:   Mother or Father of Passenger Aboard Titanic  
Child:    Son, Daughter, Stepson, or Stepdaughter of Passenger Aboard Titanic  
***
Other family relatives excluded from this study include cousins, nephews/nieces, aunts/uncles, and in-laws.  Some children travelled  
only with a nanny, therefore parch=0 for them.  As well, some travelled with very close friends or neighbors in a village, however, the definitions do not support such relations.

```python
dados = pd.read_csv('../data/train.csv', index_col='PassengerId', dtype=COLUMNS_DTYPE)
```

Cria-se uma namedtuple representando os atributos, a fim de coletar informações deles; além de enums representando os valores categóricos.

```python
def enum(**enums):
    return type('Enum', (), enums)

Atributo = namedtuple('Atributo', 'nome tipo missing desc obs')
atributos = {}

Tipo = enum(CATEGORICO='categorico', INTEIRO='inteiro', DATA='data', TIME='time', DATATIME='datatime', STRING='string')
Missing = enum(TRUE='true', FALSE='false')
```

## Scatter 2 colunas

```python
columns = ['Age', 'SibSp', 'Pclass', 'Parch', 'Fare']
columns_2a2 = [(columns[i], columns[j]) for i in xrange(len(columns) - 1) for j in xrange(i+1, len(columns))]
```

```python
fig, axs = pl.subplots(len(columns_2a2)/3, 3, figsize=(14, 14))
for ax, (x,y) in zip(axs.ravel(), columns_2a2):
    
    ax.scatter(dados[x], dados[y], c=dados.Survived)
    ax.set_title('{} x {}'.format(x, y))
```

```python
fig, axes = pl.subplots(1, 2, figsize=(16, 8))

h_rico = dados.Survived[dados.Sex == 'male'][dados.Pclass != 3].value_counts()
h_pobre = dados.Survived[dados.Sex == 'male'][dados.Pclass == 3].value_counts()

h = pd.DataFrame({'rico': h_rico, 'pobre': h_pobre})
h.plot(kind='barh', xlim=(0, 300), ax=axes[0], grid=False)
axes[0].set_title('Homem')

f_rico = dados.Survived[dados.Sex == 'female'][dados.Pclass != 3].value_counts()
f_pobre = dados.Survived[dados.Sex == 'female'][dados.Pclass == 3].value_counts()

# f = pd.DataFrame({'rica': f_rico, 'pobre': f_pobre})
pd.concat([f_rico, f_pobre], keys=['rica', 'pobra']).plot(kind='barh', xlim=(0, 300), ax=axes[1], grid=False)
axes[1].set_title('Mulher')
```

```python
dados[dados.Sex == 'male'].groupby('Survived').size().plot(kind='barh', alpha=.5, grid=False)
dados[dados.Sex == 'female'].groupby('Survived').size().plot(kind='barh', color='red', alpha=.5, grid=False)
```

```python
fig, axes = pl.subplots(1, 2, figsize=(20, 5))

death_counts = pd.crosstab([dados.Pclass, dados.Sex], dados.Survived.astype(bool))
death_counts.plot(kind='bar', stacked=True, grid=False, color=['red', 'green'], ax=axes[0])
death_counts.div(death_counts.sum(1).astype(float), axis=0).plot(kind='barh', stacked=True, color=['red', 'green'], ax=axes[1])
```
