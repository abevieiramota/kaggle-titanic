# New features

```python
transformations = []
```

## FamilySize = SibSp + Parch + 1

```python
def nf_FamilySize(df):
    
    df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

transformations.append(nf_FamilySize)
```

```python
nf_FamilySize(df)
```

```python
df.groupby(["FamilySize", "Survived"]).size().unstack().plot(kind='bar', color=['r', 'b'], figsize=(8, 6))
```

## FamilyName, FirstName, Title

```python
def nf_ExtractNames(df):
    
    df[['FamilyName', 'Title', 'FirstName']] = df["Name"].str.extract(r"(?P<FamilyName>[A-Za-z \-']+),\s(?P<Title>[A-Za-z ]+)\.\s(?P<FirstName>.*)")
    
transformations.append(nf_ExtractNames)
```

```python
nf_ExtractNames(df)
```

### analisando FamilyName e Pclass -> premissa de que se pertence à mesma família, então tem mesmo FamilyName E Pclass

FamilyNamePclass

```python
def nf_FamilyNamePclass(df):
    
    df["FamilyNamePclass"] = df.apply(lambda x: "{0}_{1}".format(x["FamilyName"], x["Pclass"]), axis=1)

transformations.append(nf_FamilyNamePclass)
```

```python
nf_FamilyNamePclass(df)
```

NSurvivedPerFamilyNamePclass -> dentro da família(FamilyNamePclass), quantos sobreviveram 

NDiedPerFamilyNamePclass -> dentro da família(FamilyNamePclass), quantos quantos morreram

DiffSurvivedDiedPerFamilyNamePclass -> dentro da família(FamilyNamePclass), quantos sobreviveram - quantos morreram

```python
gbFamilyNamePclassAndSurvived = df.\
groupby(["FamilyNamePclass", "Survived"]).\
size().\
unstack(1).\
fillna(0)
```

```python
gbFamilyNamePclassAndSurvived["1 - 0"] = gbFamilyNamePclassAndSurvived.apply(lambda x: x[1] - x[0], axis=1)
```

```python
default_gbFamilyNamePclassAndSurvived = pd.Series([0, 0, 0])
```

```python
def nf_NPerFamilyNamePclass(df):
    
    df[["NDiedPerFamilyNamePclass", "NSurvivedPerFamilyNamePclass", "DiffSurvivedDiedPerFamilyNamePclass"]] = df.\
apply(lambda x: \
      gbFamilyNamePclassAndSurvived.loc[(x['FamilyNamePclass'])] \
      
      , axis=1)

transformations.append(nf_NPerFamilyNamePclass)
```

```python
nf_NPerFamilyNamePclass(df)
```

```python
df.DiffSurvivedDiedPerFamilyNamePclass.value_counts().sort_index().plot(kind='bar', figsize=(8, 6))
```

```python
df.groupby(["NDiedPerFamilyNamePclass", "Survived"]).size().unstack().plot(kind='bar')
```

NOtherSurvivedPerFamilyNamePclass -> dentro da família(FamilyNamePclass) quantos sobreviveram, excluindo a própria pessoa 

NOtherDiedPerFamilyNamePclass -> dentro da família(FamilyNamePclass) quantos morreram, excluindo a própria pessoa

```python
def nf_NOtherPerFamilyNamePclass(df):
    
    df["NOtherDiedPerFamilyNamePclass"] = df.\
apply(lambda x: x["NDiedPerFamilyNamePclass"] - 1 if x["Survived"] == 0 else x["NDiedPerFamilyNamePclass"], axis=1)
    df["NOtherSurvivedPerFamilyNamePclass"] = df.\
apply(lambda x: x["NSurvivedPerFamilyNamePclass"] - 1 if x["Survived"] == 0 else x["NSurvivedPerFamilyNamePclass"], axis=1)

transformations.append(nf_NOtherPerFamilyNamePclass)
```

```python
nf_NOtherPerFamilyNamePclass(df)
```

```python
df.groupby(["NOtherDiedPerFamilyNamePclass", "Survived"]).size().unstack().sort_index().plot(kind='bar')
```

```python
df.groupby(["NSurvivedPerFamilyNamePclass", "Survived"]).size().unstack().sort_index().plot(kind='bar')
```

# Imputation

```python
df.isnull().sum()
```

## AgeImputedByPclassAndSex = group by (Pclass, Sex) mean(Age)

```python
meanAgeByPclassAndSex = df.groupby(["Pclass", "Sex"])["Age"].mean()
```

```python
def imp_AgeByPclassAndSex(df) :
    
    df["AgeImputedByPclassAndSex"] = df.apply(lambda x: \
                                          meanAgeByPclassAndSex.loc[(x['Pclass'], x['Sex'])] \
                                          if pd.isnull(x['Age']) \
                                          else x['Age'],\
                                          axis=1)
    
transformations.append(imp_AgeByPclassAndSex)
```

```python
imp_AgeByPclassAndSex(df)
```

```python
plt.figure(figsize=(12, 6))
plt.hist([df[df.Survived == 0]["AgeImputedByPclassAndSex"], df[df.Survived == 1]["AgeImputedByPclassAndSex"]],\
         range=(0, 90),\
         stacked=True,\
         bins=50,\
         color=['r', 'g'],\
         label=['Died', 'Survived'])
plt.legend()
```

## EmbarkedImputedByPclassAndSex = group by ()

```python
sizeByPclassAndSexAndEmbarked = df.groupby(["Pclass", "Sex", "Embarked"]).size()
```

```python
def imp_EmbarkedByClassAndSex(df):
    
    df["EmbarkedImputedByPclassAndSex"] = df.apply(lambda x: \
                                          sizeByPclassAndSexAndEmbarked.loc[(x['Pclass'], x['Sex'])].idxmax() \
                                          if pd.isnull(x['Embarked']) \
                                          else x['Embarked'],\
                                          axis=1)
transformations.append(imp_EmbarkedByClassAndSex)
```

```python
imp_EmbarkedByClassAndSex(df)
```

# More new features
## based on imputed features

```python
def nf_AgeClasses(df):
    
    df["AgeClassPerCut4Bins"] = pd.cut(df.AgeImputedByPclassAndSex, bins=4)
    df["AgeClassPerQCut4Quartiles"] = pd.qcut(df.AgeImputedByPclassAndSex, q=4)
    
    faixa_categories = ['criancinha', 'adolescente', 'adulto', 'velho']

    df.loc[df.AgeImputedByPclassAndSex < 10, "AgeClassPerFaixaEtaria"] = faixa_categories[0]
    df.loc[df.AgeImputedByPclassAndSex.between(10, 18), "AgeClassPerFaixaEtaria"] = faixa_categories[1]
    df.loc[df.AgeImputedByPclassAndSex.between(18, 50), "AgeClassPerFaixaEtaria"] = faixa_categories[2]
    df.loc[df.AgeImputedByPclassAndSex > 50, "AgeClassPerFaixaEtaria"] = faixa_categories[3]

    df.loc[:, "AgeClassPerFaixaEtaria"] = df["AgeClassPerFaixaEtaria"].astype('category')\
    .cat.set_categories(faixa_categories, ordered=True)
    
    chute_categories = ['peso', 'socooorrroooo', 'superman', 'old hero']

    df.loc[df.AgeImputedByPclassAndSex < 5, "AgeClassPerChuto"] = chute_categories[0]
    df.loc[df.AgeImputedByPclassAndSex.between(5, 15), "AgeClassPerChuto"] = chute_categories[1]
    df.loc[df.AgeImputedByPclassAndSex.between(15, 40), "AgeClassPerChuto"] = chute_categories[2]
    df.loc[df.AgeImputedByPclassAndSex > 40, "AgeClassPerChuto"] = chute_categories[3]

    df.loc[:, "AgeClassPerChuto"] = df["AgeClassPerChuto"].astype('category')\
    .cat.set_categories(chute_categories, ordered=True)
    
transformations.append(nf_AgeClasses)
```

```python
nf_AgeClasses(df)
```

```python
class Transform(object):
    
    def fit(self, X, y=None):
        
        return self
    
    def transform(self, X):
        
        for transformation in transformations:
            
            transformation(X)
            
        return X
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[('new features', Transform())])
```

```python
train = pd.read_csv('../data/train.csv')
test = pd.read_csv('../data/test.csv')
```

```python
train_final = pipe.fit_transform(train)
```

```python
test_final = pipe.fit_transform(test)
```
